````python
products = ["tshirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
  inventory = {}
  inventory = {product: int(input(f"Enter the quantity of {product}s available: ")) for product in products}
  return inventory

inventory = initialize_inventory(products)
````

Modify the function get_customer_orders so it prompts the user to enter the number of customer orders and gathers the product names using a loop and user input. Use comprehension. (Question - comprehension?)

````python
def get_customer_orders():
  customer_orders = set()
  number_customer_orders = int(input("Enter the number of customer orders: "))

  for _ in range(number_customer_orders):
    order = input("Enter the name of a product that a customer wants to order: ")
    if order in products:
      customer_orders.add(order)
  return customer_orders

customer_orders = get_customer_orders()
````

Modify the update_inventory function to remove the product from the inventory if its quantity becomes zero after fulfilling the customer orders. Use comprehension to filter out the products with a quantity of zero from the inventory.

````python
def update_inventory(customer_orders, inventory):
  updated_inventory = {product: inventory[product] - 1 if product in customer_orders else inventory[product] for product in inventory}
  updated_inventory = {product: count for product, count in updated_inventory.items() if count > 0}
  return updated_inventory

updated_inventory = update_inventory(customer_orders, inventory)

update_inventory(customer_orders, inventory)
````

````python
def calculate_order_statistics(customer_orders, products):
  total_products_ordered = len(customer_orders)
  inventory_total_items = sum(updated_inventory.values())
  percentage_of_unique_products_ordered = (total_products_ordered / inventory_total_items) * 100
  order_statistics = (total_products_ordered, percentage_of_unique_products_ordered)
  return order_statistics

order_statistics = calculate_order_statistics(customer_orders, products)
````

````python
def print_order_statistics(order_statistics):
    print("Order Statistics:")
    print("The number of products ordered is", order_statistics[0])
    print("The percentage of unique products ordered is", order_statistics[1])

print_order_statistics(order_statistics)
````

````python
def print_updated_inventory(updated_inventory):
  print("Updated Inventory: ")
  for product, quantity in updated_inventory.items():
    print(f"{product} : {quantity}")

print_updated_inventory(updated_inventory)
````

Add a new function to calculate the total price of the customer order. For each product in customer_orders, prompt the user to enter the price of that product. Use comprehension to calculate the total price. Note: assume that the user can only have 1 unit of each product.

````python
def total_price_customer_order(customer_orders):
    prices = sum(int(input(f"Enter the price of {product}: ")) for product in customer_orders)
    total_price = print("Total Price:", prices)
    return total_price

total_price_customer_order(customer_orders)
````
